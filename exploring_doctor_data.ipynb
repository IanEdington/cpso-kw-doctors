```python
doctors_df = get_geolocation_for_addresses.get_doctors_with_lat_long()
doctors_df['name'] = doctors_df['first_name'] + ' ' + doctors_df['last_name']
doctors_df['rounded_lat_lng'] = doctors_df['lat'].round(decimals=3).astype(str) + ',' + \
                                 doctors_df['lng'].round(decimals=3).astype(str)
```

```python
doctors_grouped_by_location = doctors_df.groupby('rounded_lat_lng', as_index=False)
doctors_grouped_by_location.describe()
```

```python
offices_by_3_dec = doctors_grouped_by_location.agg({'cpso':'size', 'name': ', '.join, 'location': 'max'})
offices_by_3_dec
```

```python
offices_6_plus = offices_by_3_dec[offices_by_3_dec['cpso'] > 5]
offices_2_to_5 = offices_by_3_dec[(1 < offices_by_3_dec['cpso']) &  (offices_by_3_dec['cpso'] <= 5)]
offices_1 = offices_by_3_dec[offices_by_3_dec['cpso'] == 1]
```

```python
offices_6_plus.to_csv(utils.DATA_DIR / f'offices_6_plus_rounded_3.csv')
offices_2_to_5.to_csv(utils.DATA_DIR / f'offices_2_to_5_rounded_3.csv')
offices_1.to_csv(utils.DATA_DIR / f'offices_1_rounded_3.csv')
```

# round to 4 digits since first attempt resulted in large groups

```python
ROUND_BY = 4

doctors_df = get_geolocation_for_addresses.get_doctors_with_lat_long()
doctors_df['name'] = doctors_df['first_name'] + ' ' + doctors_df['last_name']
doctors_df['rounded_lat_lng'] = doctors_df['lat'].round(decimals=ROUND_BY).astype(str) + ',' + \
                                 doctors_df['lng'].round(decimals=ROUND_BY).astype(str)

doctors_grouped_by_location = doctors_df.groupby('rounded_lat_lng', as_index=False)
doctors_grouped_by_location.describe()

offices_by_4_dec = doctors_grouped_by_location.agg({'cpso':'size', 'name': ', '.join, 'location': 'max'})

offices_6_plus = offices_by_4_dec[offices_by_4_dec['cpso'] > 5]
offices_2_to_5 = offices_by_4_dec[(1 < offices_by_4_dec['cpso']) &  (offices_by_4_dec['cpso'] <= 5)]
offices_1 = offices_by_4_dec[offices_by_4_dec['cpso'] == 1]

offices_6_plus.to_csv(utils.DATA_DIR / f'offices_6_plus_rounded_4.csv')
offices_2_to_5.to_csv(utils.DATA_DIR / f'offices_2_to_5_rounded_4.csv')
offices_1.to_csv(utils.DATA_DIR / f'offices_1_rounded_4.csv')
```
